1. Find the angle between the curves r = 4(1+cos(t)) and r = 5(1-cos(t))

```python
from sympy import *
r,t = symbols ('r,t')               # Define the variables required as symbols
r1=4*(1+cos (t));                  # Input first polar curve
r2=5*(1-cos (t));                 # Input first polar curve
dr1 = diff (r1 ,t)               # find the derivative of first function
dr2 = diff (r2 ,t)              # find the derivative of secodn function
t1=r1/dr1
t2=r2/dr2
q= solve (r1-r2 ,t)         # solve r1 ==r2 , to find the point of intersection between curves
w1=t1.subs ({t: float (q[1])}) #substitute the value of "t" in t1
w2=t2. subs ({t: float (q[1])}) #substitute the value of "t" in t2
y1= atan (w1)                    # to find the inverse tan of w1
y2= atan (w2)                     # to find the inverse tan of w2
w=abs(y1-y2)                       # angle between two curves is abs(w1 -w2)
print ('Angle between curves in radians is %0.3f '%(w))
```

2. Find the radius of curvature r = 4(1 + cos(t)) at t=pi/2

```python
from sympy import *
t= Symbol ('t')      # define t as symbol
r= Symbol ('r')
r=4*(1+cos(t))
r1= Derivative (r,t). doit ()      # find the first derivative of r w.r.t "t"
r2= Derivative (r1 ,t). doit ()    # find the second derivative of r w.r.t "t"
rho =(r ** 2+r1 ** 2) ** (1.5)/(r ** 2+2*r1 ** 2-r*r2);
# Substitute r1 and r2 in formula
rho1 = rho . subs (t,pi/2)         # substitute t in rho
print ('The radius of curvature is %3.4f units '% rho1 )
```

Partial derivatives
The partial derivative of f(x,y) with respect to x at the point (x0,y0) is
fx=dho f/dho x at (x0,y0) = lim h-0 f(x0 + h, y0) - f(x0,y0)/ h
The partial derivative of f(x,y) with respect to y at the point (x0,y0) is
fy=dho f/dho y at (x0,y0) = lim h-0 f(x0,y0 + h) - f(x0,y0)/ h

Jacobians

Let x = g(u,v) and y = h(u,v) be a transformation of the plane. Then the Jacobian of
this transformation is  J= dho (x,y,z) / dho (u,v,w) = [dho x/dho u , dho x/dho v , dho x/dho w \n
dho y/dho u , dho y/dho v , dho y/dho w \n
dho z/dho u , dho z/dho v , dho z/dho w]

1. If u = xy/z , v = yz/x , w = zx/y then prove that J = 4

```python
from sympy import *
x,y,z= symbols ('x,y,z')
u=x*y/z
v=y*z/x
w=z*x/y
# find the all first order partial derivates
dux = diff (u,x)
duy = diff (u,y)
duz = diff (u,z)
dvx = diff (v,x)
dvy = diff (v,y)
dvz = diff (v,z)
dwx = diff (w,x)
dwy = diff (w,y)
dwz = diff (w,z)
# construct the Jacobian matrix
J= Matrix ([[dux ,duy ,duz],[dvx ,dvy ,dvz],[dwx ,dwy ,dwz]]);
print ("The Jacobian matrix is \n")
display (J)
# Find the determinat of Jacobian Matrix
Jac =det (J). doit ()
print ('\n\n J = ', Jac )
```

Maxima and minima problem

Find the Maxima and minima of f(x,y) = x^3 + xy^2 - 15x^2 - 15y^2 + 72x

```python
import sympy as sym
x,y=sym.symbols('x y')
f=x**3+3*x*y**2-15*x**2-15*y**2+72*x
dx1=sym.Derivative(f,x).doit()
dy1=sym.Derivative(f,y).doit()
eq1,eq2=sym.Eq(dx1,0),sym.Eq(dy1,0)
sol=sym.solve([eq1,eq2],(x,y))
print("the critical points are",sol)
dx2=sym.Derivative(f,x,2).doit()
dxy=sym.Derivative(dx1,y).doit()
dy2=sym.Derivative(f,y,2).doit()
for i,j in sol:
    A=dx2.subs({x:i,y:j}).evalf()
    B=dxy.subs({x:i,y:j}).evalf()
    C=dy2.subs({x:i,y:j}).evalf()
    dis=A*C-B*B
if(dis>0 and A>0):
    print("The function as minimum at(",i,",",j,")")
    q1=f.subs({x:i,y:j}).evalf()
    print("Minimum value:",q1)
elif(dis>0 and A<0):
    print("The function as maximum at(",i,",",j,")")
    q2=f.subs({x:i,y:j}).evalf()
    print("Maximum value:",q2)
elif(dis<0):
    print("(",i,",",j,") is a saddle point")
elif(dis==0):
    print("It needs furthur investigation")
```

3. Find the Maclaurin series expansion of sin(x) + cos(x) upto 3rd degree term. Calculate sin(10) + cos(10).

```python
import numpy as np
from matplotlib import pyplot as plt
from sympy import *

x = Symbol('x')
y = sin(x) + cos(x)

x0 = float(0)
dy = diff(y, x)
d2y = diff(y, x, 2)
d3y = diff(y, x, 3)

yat = lambdify(x, y)
dyat = lambdify(x, dy)
d2yat = lambdify(x, d2y)
d3yat = lambdify(x, d3y)

y = yat(x0) + ((x - x0) / 1) * dyat(x0) + ((x - x0)**2 / 2) * d2yat(x0) + ((x - x0)**3 / 6) * d3yat(x0)

print(simplify(y))

yat = lambdify(x, y)
print("%.3f" % yat(10*(np.pi/180)))

def f(x):
    return np.sin(x) + np.cos(x)

x_vals = np.linspace(-10, 10)
plt.plot(x_vals, yat(x_vals), color='red')
plt.plot(x_vals, f(x_vals), color='green')
plt.ylim([-3, 3])
plt.title('Maclaurin series by Hemanth S.P')
plt.grid()
plt.show()
```

Example 5: Obtain the solution of 3x + 5y = 1; x + y = 1 graphically.

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
x, y = symbols('x y')
# Solve the system of equations
sol = solve([3*x + 5*y - 1, x + y - 1], [x, y])
p = sol[x]
q = sol[y]
print('Point of intersection is A (', p, ',', q, ')\n')
# Define x range
x_vals = np.arange(-10, 10, 0.001)
# Equations of the lines
y1 = (1 - 3 * x_vals) / 5
y2 = 1 - x_vals
# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1, label='$3x+5y=1$')
plt.plot(x_vals, y2, label='$x+y=1$')
# Plotting the point of intersection
plt.plot(p, q, 'ro')  # 'ro' -> red color, circle marker
plt.text(p + 0.2, q, 'A(' + str(p.evalf()) + ', ' + str(q.evalf()) + ')')
# Axes and labels
plt.axhline(y=0, color='black',linewidth=0.5)
plt.axvline(x=0, color='black',linewidth=0.5)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('Values of x')
plt.ylabel('Values ofy')
plt.title('$3x+5y=1; x+y=1$')
plt.legend()
plt.grid()
plt.show()
```

LAB 9: Solution of system of linear equations by Gauss-Seidel method.

Example 1: Solve the system of equations using Gauss-Seidel method: 20x + y - 2z = 17 , 3x + 20y - z = - 18 , 2x - 3y + 20z = 25

```python
f1 = lambda x, y, z: (17 - y + 2 * z) / 20
f2 = lambda x, y, z: (-18 - 3 * x + z) / 20
f3 = lambda x, y, z: (25 - 2 * x + 3 * y) / 20
x0 = 0
y0 = 0
z0 = 0
count = 1
e = float(input('Enter tolerable error: '))
# Implementation of Gauss Seidel Iteration
print('\nCount \tx\ty\tz\n')
condition = True
while condition:
    x1 = f1(x0, y0, z0)
    y1 = f2(x1, y0, z0)
    z1 = f3(x1, y1, z0)
    print('%d\t%0.4f\t%0.4f\t%0.4f\n' % (count, x1, y1, z1))
    e1 = abs(x0 - x1)
    e2 = abs(y0 - y1)
    e3 = abs(z0 - z1)
    count += 1
    x0 = x1
    y0 = y1
    z0 = z1
    condition = e1 > e and e2 > e and e3 > e
print('\nSolution: x=%0.3f y=%0.3f and z=%0.3f\n' % (x1, y1, z1))
```
